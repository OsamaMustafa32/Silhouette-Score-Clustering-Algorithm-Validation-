```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_blobs(n_samples=500,
                 n_features=2,
                 centers=4,
                 cluster_std=1,
                 center_box=(-10.0,10.0),
                 shuffle=True,
                 random_state=1)
```

```python
print(X.shape)
print(y.shape)
```

**make_blobs** function to generate data with 4 clusters. Each cluster will have its own randomly generated centroid or center around which the data points for that cluster will be distributed.

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, edgecolors='k')
plt.title('Sample Data from make_blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from enum import auto
wcss=[]

for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0,n_init='auto')
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### Hence From the Elbow Method we can take number of cluster(k) = 4
